```python
df2['사용시간(hr)']=df2['사용시간(분)']/60
df2['사용시간(hr)']=round(df2['사용시간(hr)'],2)
df2
```

```python
del df2['사용시간(분)']
df2
```

```python
df2['속도(km/h)']=df2['이동거리(KM)']/df2['사용시간(hr)']
df2['속도(km/h)']=round(df2['속도(km/h)'],2)
df2
```
