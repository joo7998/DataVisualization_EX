```python
import pandas as pd
from pandas import DataFrame
```

```python
df = pd.read_csv('bicycle4.csv', encoding='utf-8')
df
```

```python
del df['Unnamed: 0']
df
```

```python
df.rename(columns={'대여시간':'대여시간(24)'},inplace=True)
df
```

```python
# 월 연도 추출 
import numpy as np
import datetime
```

```python
# 데이터의 원본을 보존하기 위해 데이터를 복사
pandas_copy = df.copy()
```

```python
pandas_copy
```

```python
# pandas.Series.dt.year() & pandas.Series.dt.month()의 object - Datetime 항목에 대해 
# np 연도 및 월의 배열을 리턴
dates = pandas_copy['대여일자']
pandas_copy = pd.DataFrame({'대여일자': pd.to_datetime(dates)})
```

```python
pandas_copy['년'] = pandas_copy['대여일자'].dt.year
pandas_copy['월'] = pandas_copy['대여일자'].dt.month
pandas_copy
```

```python
# df.join(pandas_copy, how='left')
df2 = pd.merge(df, pandas_copy, left_index=True, right_index=True, how='left')
df2
```

```python
df2.rename(columns={'대여일자_x':'대여일자'},inplace=True)
del df2['대여일자_y']
df2
```

```python
# 1. 년도 별
# df2['년'].unique() 
df2['년'].value_counts(ascending=True, dropna=False)
                                                        # 2017     97002
                                                        # 2018    148884
                                                        # 2020    157664
                                                        # 2019    297923
# df2.sort_values(by='년', ascending=True)


# df['column_name'].value_counts()[:20]
# df2['월'].value_counts(sort=False, ascending=True, dropna=False)                 유일값 오름차
#                       (sort=True, ascending=False, dropna=False)                 갯수 많은 순서대로 내림차 

#                                             default : value_counts(sort=True, ascending=False) 
                                           
#                                             sort=True : 개수 기준 정렬,           False 유일한 값 기준 
#                                             ascending=False : 내림차,            True 오름차순 
#                                             dropna=False : NaN 포함
```
